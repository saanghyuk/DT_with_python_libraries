```python
import pandas as pd
import os
encoding = 'latin1'
```

```python
upath=os.path.expanduser('/Users/sanghyuk/Documents/data_analysis_with_python/pydata-book/datasets/movielens/users.dat')
rpath=os.path.expanduser('/Users/sanghyuk/Documents/data_analysis_with_python/pydata-book/datasets/movielens/ratings.dat')
mpath= os.path.expanduser('/Users/sanghyuk/Documents/data_analysis_with_python/pydata-book/datasets/movielens/movies.dat')
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
mnames = ['movie_id', 'title', 'genres']
```

```python
users=pd.read_csv(upath, sep='::', header=None, names=unames, encoding = encoding)
ratings=pd.read_csv(rpath, sep='::', header=None, names=rnames, encoding = encoding)
movies=pd.read_csv(mpath, sep='::', header=None, names=mnames, encoding = encoding)
```

```python
users[::5]
movies[::5]
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data
```

```python
data.iloc[0]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
```

```python
mean_ratings
```

```python
ratings_by_title = data.groupby('title').size()
```

```python
data.head()
```

```python
ratings_by_title
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
```

```python
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
```

```python
top_female_ratings
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```
